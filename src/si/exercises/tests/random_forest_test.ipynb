```python
from si.io.csv_file import read_csv          
from si.model_selection.split import train_test_split
from si.models.random_forest_classifier import RandomForestClassifier

def main():
    # 1) Carregar o dataset iris
    dataset = read_csv("datasets/iris.csv", sep=",", features=True, label=True)

    # 2) Dividir em train e test
    train_ds, test_ds = train_test_split(dataset, test_size=0.2, random_state=42)

    # 3) Criar o modelo RandomForestClassifier
    rf = RandomForestClassifier(
        n_estimators=50,     # número de árvores
        max_features=None,   # se None, usa sqrt(n_features)
        min_sample_split=2,
        max_depth=None,
        mode="gini",
        seed=42
    )

    # 4) Treinar o modelo
    rf._fit(train_ds)

    # 5) Avaliar no conjunto de teste
    score = rf._score(test_ds)

    print(f"RandomForest accuracy on iris test set: {score:.3f}")


if __name__ == "__main__":
    main()
```
