```python
import numpy as np
from si.decomposition.pca import PCA
```

```python
np.random.seed(0)
X = np.random.rand(10, 4)  # Example data
pca = PCA(n_components=2)
pca._fit(X)
X_reduced = pca._transform(X)

print(X_reduced[:5])  # Reduced data
print(pca.explained_variance)  # Explained variance
```
