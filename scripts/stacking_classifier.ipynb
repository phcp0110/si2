Carregar o dataset

```python
from si.io.csv_file import read_csv

dataset = read_csv(
    "datasets/breast-bin.csv",
    sep=",",
    features=True,
    label=True
)

dataset
```

Dividir em treino e teste

```python
from si.model_selection.split import train_test_split

train_ds, test_ds = train_test_split(
    dataset,
    test_size=0.2,
    random_state=42
)
```

Criar um modelo KNNClassifier

```python
from si.models.knn_classifier import KNNClassifier

knn_1 = KNNClassifier(k=3)
```

Criar um modelo LogisticRegression

```python
from si.models.logistic_regression import LogisticRegression

log_reg = LogisticRegression(lr=0.01, max_iter=1000)
```

Criar um modelo DecisionTree

```python
from si.models.decision_tree_classifier import DecisionTreeClassifier

tree = DecisionTreeClassifier(
    max_depth=5,
    min_sample_split=2,
    mode="gini"
)
```

Criar o modelo final

```python
knn_final = KNNClassifier(k=5)
```

Criar o StackingClassifier

```python
from si.ensemble.stacking_classifier import StackingClassifier

stacking = StackingClassifier(
    models=[knn_1, log_reg, tree],
    final_model=knn_final
)
```

Treinar e avaliar o modelo

```python
stacking._fit(train_ds)

score = stacking._score(test_ds)
print(f"StackingClassifier accuracy: {score:.4f}")
```
