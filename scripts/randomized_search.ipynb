Dataset breast-bin.csv

```python
import sys
import os
sys.path.append(os.path.abspath("../src"))

from si.io.csv_file import read_csv

dataset = read_csv(
    "datasets/breast-bin.csv",
    sep=",",
    features=True,
    label=True
)
```

LogisticRegression

```python
from si.models.logistic_regression import LogisticRegression

model = LogisticRegression()
```

Definir distribuições de hiperparâmetros

```python
import numpy as np

param_dist = {
    "l2_penalty": np.linspace(1, 10, 10),
    "alpha": np.linspace(0.001, 0.0001, 100),
    "max_iter": np.linspace(1000, 2000, 200).astype(int),
}
```

Executar randomized search

```python
from si.model_selection.randomized_search import randomized_search_cv
from si.metrics.accuracy import accuracy

results = randomized_search_cv(
    model=model,
    dataset=dataset,
    hyperparameter_grid=param_dist,
    scoring=accuracy,
    cv=3,
    n_iter=10,
    random_state=42
)
```
